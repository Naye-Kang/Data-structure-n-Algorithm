```python
texts = [' 우리나라    대한민국, 우리나라%$ 만세', '비아그&라 500GRAM 정력 최고!', '나는 대한민국 사람', '보험료 15000원에 평생 보장 마감 임박', '나는 홍길동']
```

```python
def clean_text(text):
    text_re1=[t.lower() for t in text]
    text_re2=[sub("[0-9]",'',text) for text in text_re1]
    text_re3=[sub("[,.?!:;]",'',text) for text in text_re2]
    text_re4=[sub(spec_str,'',text) for text in text_re3]
    text_re5=[''.join(findall("[^a-z]",text)) for text in text_re4]
    text_re6=[' '.join(text.split()) for text in text_re5]
    text_re7=[text.split() for text in text_re5]
    return text_re7
```

```python
clean_text(texts)
```

```python
texts_result=clean_text(texts[0])
```
